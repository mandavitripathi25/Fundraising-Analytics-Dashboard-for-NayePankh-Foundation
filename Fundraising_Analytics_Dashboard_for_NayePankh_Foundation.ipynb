**Create Sample Dataset**

```python
import pandas as pd

data = {
    'Month':['Jan','Feb','Mar','Apr','May','Jun'],
    'Donations':[50000,65000,70000,85000,90000,120000],
    'Donors':[120,145,160,180,200,250],
    'Volunteers':[45,50,52,60,65,72]
}

df = pd.DataFrame(data)
df
```

**Data Exploration**

```python
print(df.info())

print(df.describe())
```

**Monthly Donation Trend**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Donations'], marker='o')
plt.title('Monthly Donation Trend')
plt.xlabel('Month')
plt.ylabel('Donations (₹)')
plt.grid(True)
plt.show()
```

**Donor Growth Analysis**

```python
plt.figure(figsize=(8,5))
plt.bar(df['Month'], df['Donors'], color='green')
plt.title('Monthly Donor Growth')
plt.xlabel('Month')
plt.ylabel('Number of Donors')
plt.show()
```

**Volunteer Engagement Analysis**

```python
plt.figure(figsize=(8,5))
plt.bar(df['Month'], df['Volunteers'], color='orange')
plt.title('Volunteer Participation Trend')
plt.xlabel('Month')
plt.ylabel('Volunteers')
plt.show()
```

**KPI Dashboard**

```python
total_donations = df['Donations'].sum()
total_donors = df['Donors'].sum()
avg_donation = total_donations/total_donors

print("Total Donations: ₹", total_donations)
print("Total Donors:", total_donors)
print("Average Donation per Donor: ₹", round(avg_donation,2))
```

**Automated Report Generation**

```python
report = f"""
NayePankh Foundation Analytics Report

Total Donations: ₹{total_donations}
Total Donors: {total_donors}
Average Donation: ₹{round(avg_donation,2)}

Highest Donation Month:
{df.loc[df['Donations'].idxmax(),'Month']}
"""

print(report)
```
